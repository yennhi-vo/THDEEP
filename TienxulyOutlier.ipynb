```
import pandas as pd
import numpy as np


# Nạp dữ liệu
url = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/titanic_disaster.csv'
titanic = pd.read_csv(url)

# Xử lý giá trị thiếu cho cột Age bằng trung vị (median) để tránh ảnh hưởng đến việc tính toán outlier
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())

# Hiển thị thống kê mô tả để xem sơ bộ về phân phối
print(titanic[['Age', 'Fare']].describe())
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Đếm số lượng outlier
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Số lượng outlier trong cột {column}: {len(outliers)}")

    # Có 2 cách xử lý: Loại bỏ hoặc Thay thế (Capping)
    # Ở đây chúng ta chọn cách loại bỏ để dữ liệu sạch hơn
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df_cleaned

# Thực hiện xử lý cho cột Fare (Giá vé thường có outlier rất lớn)
titanic_cleaned = handle_outliers_iqr(titanic, 'Fare')

# Tiếp tục xử lý cho cột Age (Tuổi) trên tập dữ liệu đã lọc Fare
titanic_final = handle_outliers_iqr(titanic_cleaned, 'Age')

print(f"\nKích thước tập dữ liệu gốc: {titanic.shape}")
print(f"Kích thước tập dữ liệu sau khi xóa outlier: {titanic_final.shape}")
```
